```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import gc
from ES import ES
from vritual_machine import VM
```

```python
noise = 0.1
dim = 3
```

```python
def bechmark(dim,
             budget = 1000,
             ave_time = 2,
             auto_kES = False,
             auto_kES_kwargs = None,
             auto_aES = False,
             auto_aES_kwargs = None,
             noise = noise,
             vm=None,
             es=None,):
    
    aES = 0.001*np.ones(dim)
    x_bounds = ((-2,2),)*dim
    
    if vm is None:
        vm = VM(x0 = 2*np.ones(dim),
            decision_CSETs=['COR'+str(i) for i in range(dim)],
            objective_RDs =['Loss'],
            objective_RD_noises = [noise])
    
    def loss_func(x,time_span=ave_time,vm=vm):
        vm.ensure_set(vm.decision_CSETs,
                      None,
                      x)
        ave_data, raw_data = vm.fetch_data(vm.objective_RDs,time_span=time_span,abs_z=3.)
        return ave_data['Loss']

    if es is None:
        es = ES(loss_func=loss_func,
                aES = aES,
                x0 = 2*np.ones(dim),
                x_bounds=x_bounds,
               )
        
    if budget>0:
        es(
            int(budget/ave_time),
            auto_kES = auto_kES,
            auto_kES_kwargs = auto_kES_kwargs,
            auto_aES = auto_aES,
            auto_aES_kwargs = auto_aES_kwargs,
           )

        VMhist = pd.DataFrame(vm.history)
        fig,ax = plt.subplots(1,2,figsize=(10,2))
        VMhist.plot('t',vm.decision_CSETs,ax=ax[0],alpha=0.7)
        VMhist.plot('t','Loss',ax=ax[1],alpha=0.7)

        EShist = pd.DataFrame(es.history)
        fig,ax = plt.subplots(1,2,figsize=(8,2))
        EShist.plot('t',['aES'+str(i) for i in range(2)],ax=ax[0],alpha=0.7)
        EShist.plot('t',['kES'+str(i) for i in range(2)],ax=ax[1],alpha=0.7)
    
    return vm,es
```

```python
# dim=2
n=dim*64

vm,es = bechmark(
             dim = dim,
             budget = 0,
             ave_time=0.4,
             auto_kES = False,
             auto_aES = False,)

es(n,kES=np.zeros(es.dim),aES = 0.001*np.ones(es.dim))
y = np.array(es.history['y'][1:])
y -= y.mean()
fft = np.abs(np.fft.rfft(y))/n
nu = 0.5*np.arange(len(fft))/len(fft)
dft = []
T = np.arange(n)
for _ in nu:
    dft.append((np.sum(y*np.sin(2*np.pi*_*T))**2 + np.sum(y*np.cos(2*np.pi*_*T))**2)**0.5/n)
dft = np.array(dft)
plt.figure(figsize=(15,2))
for i in range(dim):
    plt.vlines(es.wES[i]/np.pi/2,fft.min(),fft.max(),color='r',alpha=0.5)
plt.plot(nu,fft,alpha=0.9)
plt.plot(nu,dft,alpha=0.9)
wES = es.wES/(2*np.pi)
```
