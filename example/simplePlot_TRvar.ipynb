# How TR_var penalize acqusition

```python
var = np.linspace(0,3,100)
def UCB(beta,TR_var=0):
    f = (beta*var)**0.5
    if TR_var > 0:
        f-=(beta/TR_var)**0.5*var
    return  f 

fig,ax = plt.subplots(1,4,figsize=(15,2),dpi=96)
for beta in [1,4,9]:
    TR_var = 0
    ax[0].plot(var,UCB(beta,TR_var),label=f'b{beta}t{TR_var}')
ax[0].legend()

for beta in [1,4,9]:
    TR_var = 3
    ax[1].plot(var,UCB(beta,TR_var),label=f'b{beta}t{TR_var}')
ax[1].legend()

for beta in [1,4,9]:
    TR_var = 2
    ax[2].plot(var,UCB(beta,TR_var),label=f'b{beta}t{TR_var}')
ax[2].legend()

for beta in [1,4,9]:
    TR_var = 1
    ax[3].plot(var,UCB(beta,TR_var),label=f'b{beta}t{TR_var}')
ax[3].legend()
```
